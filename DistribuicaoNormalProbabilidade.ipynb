```python
#Importação de pacotes

import pandas as pd
import seaborn as sns
from scipy.stats import norm
import numpy as np
import statistics as st
import matplotlib.pyplot as fig
```

```python
#Carregando o conjunto de dados
df = pd.read_csv('alturas.csv')
```

```python
#Visualizando o dataframe
df.head()
```

```python
#Explorando algumas medidas estatisticas
df.describe()
```

```python
#Extraindo min, max, media e desvio padrão
vlMin = df.min()
vlMax = df.max()
media = st.mean(df['altura'])
desvio_padrao = st.pstdev(df['altura'])
print(f'Valor maximo: {vlMax}\nValor minimo: {vlMin}\nmédia: {media}\nDp: {desvio_padrao}')
```

```python
#Histograma com a curva de distribuição
sns.histplot(df['altura'], kde = True, discrete = True)
```

```python
#Mostrar somente a Curva da distribuição
sns.kdeplot(data = df)
```

```python
#Gerando outros gráficos

vlMin = df.min()
vlMax = df.max()
media = st.mean(df['altura'])
desvio_padrao = st.pstdev(df['altura'])

fig.subplot(211)
fig.plot(df['altura'], '-b')

fig.subplot(212)
fig.hist(df, bins = 10, density = True)

eix = np.linspace(vlMin, vlMax,100)
eiy = norm.pdf(eix, media, desvio_padrao)
fig.plot(eix, eiy, color = 'black')
```

```python
# Obtendo a probabilidade

probabilidade = norm.cdf(175, 165, 8) - norm.cdf(155,165,8)
print('Probabilidade de: ', round(probabilidade * 100,2), '%')
```
